```python
# Display 360x360 pixel image
Image(filename = 'checkerboard.jpg')
```

```python
# Read image as gray scale
cb_img = cv2.imread('checkerboard.jpg',0)

# Print the image data as a numpy 2d array
# Each pixel value is 8 bits (0,255)
print(cb_img)
```

## Diaplaying Image attributes

```python
# Print size of image
print("Size of image: ",cb_img.shape)
print("Size of image: ",cb_img.size)

# Print data type of image
print("Data type of image: ",cb_img.dtype)
```

## Displaying image using Matplotlib

```python
# Display image
plt.figure(figsize=[2,2]) 
plt.imshow(cb_img)
```

## What happended?
Even though the image was in gray scale, it isn't necessarily show a gray scale image while using imshow() because Matplotlib use different colour maps and it is possible that the gary scale is not set.

## How to fix this issue?

```python
# Set color map to gray scale to render image in proper color using matplotlib
plt.figure(figsize=[2,2])
plt.imshow(cb_img,cmap="gray")
```

## Read and Display the color image 

```python
# Read image 
cola_img = cv2.imread('coke-logo.webp',1)

# Print image size
print('Size of image: ',cola_img.shape)
print('Size of image: ',cola_img.size)

#Print data type of image
print('Data type of image: ',cola_img.dtype)
```

## Display the image

```python
plt.figure(figsize=[5,5])
plt.imshow(cola_img)
```

## What happened?
The color is different from the original image while plotting using matplotlib is due to the fact that matplotlib expects color to be in RGB format whereas OpenCv stores image data in BGR format. So to fix this we have to revers e the channels of the image.

## How to fix this issue?

```python
cola_img_channel_reverse = cola_img[:, :, ::-1]
plt.figure(figsize=[5,5])
plt.imshow(cola_img_channel_reverse)
```

## Splitting and Merging Color Channels

```python
# Split the image in B,g,r components 
img_lake = cv2.imread('lake.jpeg',1)
b,g,r = cv2.split(img_lake)
# Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(r,cmap='gray'); plt.title('Red Channel');
plt.subplot(142); plt.imshow(g,cmap='gray'); plt.title('Green Channel');
plt.subplot(143); plt.imshow(b,cmap='gray'); plt.title('Blue Channel');

#plt.subplot(no of rows,no of colums,index). Eg plt.subplot(142) means there are 4 images in 1 row and we are placing that image in index 2 i.e in the 2nd position.


# Merge the individual channels into Bgr image
img_Merged = cv2.merge((r,g,b))

#Show merged image
plt.subplot(144); plt.imshow(img_Merged, cmap='gray'); plt.title('Merged')
```

## Changing from BGR to RGB

```python
img_lake_rgb = cv2.cvtColor(img_lake, cv2.COLOR_BGR2RGB)
plt.imshow(img_lake_rgb)
```

## Changing to HSV color space 

```python
img_hsv = cv2.cvtColor(img_lake, cv2.COLOR_BGR2HSV)

# Split the image into H,S,V components

h,s,v = cv2.split(img_hsv)

# Show the channels
plt.figure(figsize=[25,5])
plt.subplot(151); plt.imshow(h,cmap='gray'); plt.title('H Channel');
plt.subplot(152); plt.imshow(s,cmap='gray'); plt.title('S Channel');
plt.subplot(153); plt.imshow(v,cmap='gray'); plt.title('V Channel');
plt.subplot(154); plt.imshow(img_hsv,cmap='gray'); plt.title('HSV');
plt.subplot(155); plt.imshow(img_lake_rgb,cmap='gray');plt.title('Original');
```

Hue(H channel) represensts the color of the image(actual representation of the color), saturation(s channel) represents the intensity of the color(eg pure red vs a dull red), v(v channel) represents the value(how light or dark the color is).
